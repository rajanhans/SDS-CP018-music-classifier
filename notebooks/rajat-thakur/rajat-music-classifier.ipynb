# Some Terminologies:

1. Waveform:
Think of it as the "heartbeat" of the sound.

When we speak, sing, or play music, the sound goes up and down like waves in the sea.
A waveform is a picture of how loud or quiet the sound is over time.
Example: If you see a waveform with big waves, the sound is loud. If the waves are small, the sound is soft.
It tells us: How loud or quiet the sound is at different times.



2. Spectrogram:
Imagine we use a magical glass to see the hidden ingredients of the sound.

Every sound is made of many "ingredients" (frequencies). For example, a guitar string has high and low notes mixed together.


A spectrogram shows how strong each ingredient (frequency) is at different times.
It’s like seeing a rainbow of colors, where bright parts mean strong sounds and dark parts mean weak sounds.
It tells us: What kind of notes (frequencies) are in the sound and when they happen.

Mel-Spectrogram
Okay, the Mel-spectrogram is similar to a regular spectrogram, but it’s special because it focuses more on how humans hear sound.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
import random
```

```python
cnn = Sequential()

# Convolutional Layer 1
cnn.add(Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
cnn.add(BatchNormalization())  # Stabilize learning
cnn.add(MaxPool2D(pool_size=2, strides=2))  # Downsample image

# Convolutional Layer 2
cnn.add(Conv2D(filters=64, kernel_size=3, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(MaxPool2D(pool_size=2, strides=2))

# Convolutional Layer 3
cnn.add(Conv2D(filters=128, kernel_size=3, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(MaxPool2D(pool_size=2, strides=2))

# Flattening
cnn.add(Flatten())

# Fully Connected Layer
cnn.add(Dense(units=256, activation='relu'))  # Increased units for complexity
cnn.add(Dropout(0.5))  # Prevent overfitting

# Output Layer
cnn.add(Dense(units=10, activation='softmax'))  # 10 classes
```

```python
cnn.compile(
    optimizer=Adam(learning_rate=0.0001),  # Lower learning rate for stability
    loss='categorical_crossentropy',  # Ensure labels are one-hot encoded
    metrics=['accuracy']
)
```

```python
# Path to the dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_dir = "/Users/rajatthakur/MyDocumets/SuperDatascience DataSet/Music ClassifierGTZAN Dataset/newImages/train"
test_dir = "/Users/rajatthakur/MyDocumets/SuperDatascience DataSet/Music ClassifierGTZAN Dataset/newImages/test"


import numpy as np
import random
import tensorflow as tf

# Set seed for reproducibility
seed = 42
np.random.seed(seed)
random.seed(seed)
tf.random.set_seed(seed)


# Training data generator
train_datagen = ImageDataGenerator(
    rescale=1./255,         # Normalize pixel values
    shear_range=0.2,        # Shear transformations
    zoom_range=0.2,         # Random zoom
    horizontal_flip=True    # Random horizontal flip
)

# Test data generator
test_datagen = ImageDataGenerator(rescale=1./255)

# Load images from directories
training_set = train_datagen.flow_from_directory(
    train_dir,
    target_size=(64, 64),   # Resize images to match CNN input
    batch_size=32,
    class_mode='categorical'
)

test_set = test_datagen.flow_from_directory(
    test_dir,
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
import random


cnn = Sequential()

# Convolutional layer and pooling
cnn.add(Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
cnn.add(MaxPool2D(pool_size=2, strides=2))

# Add another convolutional layer and pooling
cnn.add(Conv2D(filters=64, kernel_size=3, activation='relu'))
cnn.add(MaxPool2D(pool_size=2, strides=2))

# Flattening
cnn.add(Flatten())

# Fully connected layer
cnn.add(Dense(units=128, activation='relu'))

# Output layer (Softmax for multi-class classification)
cnn.add(Dense(units=10, activation='softmax'))  # 10 units for 10 genres

# Compile the CNN
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print(cnn.summary())
```

```python
cnn.fit(
    x=training_set,
    validation_data=test_set,
    epochs=25
)
```
