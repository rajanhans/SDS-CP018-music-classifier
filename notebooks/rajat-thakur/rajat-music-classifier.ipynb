```python
import matplotlib.pyplot as plt

# Select one track
track = features_30_sec.iloc[0]

# Plot MFCC means
mfcc_means = track[['mfcc1_mean', 'mfcc2_mean', 'mfcc3_mean', 'mfcc4_mean', 'mfcc5_mean', 
                    'mfcc6_mean', 'mfcc7_mean', 'mfcc8_mean', 'mfcc9_mean', 'mfcc10_mean', 
                    'mfcc11_mean', 'mfcc12_mean', 'mfcc13_mean', 'mfcc14_mean', 'mfcc15_mean', 
                    'mfcc16_mean', 'mfcc17_mean', 'mfcc18_mean', 'mfcc19_mean', 'mfcc20_mean']]
print('MFCC (Mel-Frequency Cepstral Coefficients) Shows the tonal distribution of an audio track. It Represent the tone, timbre, and texture of the sound.')
plt.figure(figsize=(10, 6))
plt.bar(range(1, 21), mfcc_means, color='teal')
plt.title(f"MFCC Means for {track['filename']}")
plt.xlabel("MFCC Coefficients")
plt.ylabel("Mean Value")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = features_30_sec

# Define MFCC columns
mfcc_columns = [f'mfcc{i}_mean' for i in range(1, 21)]

# Group by label and calculate the mean for each label
grouped_data = data.groupby('label')[mfcc_columns].mean()

# Plot bar plots for each label
labels = grouped_data.index
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(25, 15))
axes = axes.flatten()

for i, label in enumerate(labels):
    axes[i].bar(range(1, 21), grouped_data.loc[label], color='teal')
    axes[i].set_title(f'MFCC Means: {label}', fontsize=20)
    axes[i].set_xlabel('MFCC Coefficients', fontsize=20)
    axes[i].set_ylabel('Mean Value', fontsize=20)

print('Genres with a similar tonal structure may have a similar shape across their MFCC coefficients.')
print('As shown for Classical Music: Likely to have more stable and balanced MFCC values, representing smooth transitions and harmonious tones.')
print('Metal/Rock: Shows spikes in certain MFCCs, representing the sharpness and complexity of electric instruments.')

plt.tight_layout()
plt.show()
```
