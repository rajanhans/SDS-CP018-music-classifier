```python
import matplotlib.pyplot as plt

# Select one track
track = features_30_sec.iloc[0]

# Plot MFCC means
mfcc_means = track[['mfcc1_mean', 'mfcc2_mean', 'mfcc3_mean', 'mfcc4_mean', 'mfcc5_mean', 
                    'mfcc6_mean', 'mfcc7_mean', 'mfcc8_mean', 'mfcc9_mean', 'mfcc10_mean', 
                    'mfcc11_mean', 'mfcc12_mean', 'mfcc13_mean', 'mfcc14_mean', 'mfcc15_mean', 
                    'mfcc16_mean', 'mfcc17_mean', 'mfcc18_mean', 'mfcc19_mean', 'mfcc20_mean']]
print('MFCC (Mel-Frequency Cepstral Coefficients) Shows the tonal distribution of an audio track. It Represent the tone, timbre, and texture of the sound.')
plt.figure(figsize=(10, 6))
plt.bar(range(1, 21), mfcc_means, color='teal')
plt.title(f"MFCC Means for {track['filename']}")
plt.xlabel("MFCC Coefficients")
plt.ylabel("Mean Value")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = features_30_sec

# Define MFCC columns
mfcc_columns = [f'mfcc{i}_mean' for i in range(1, 21)]

# Group by label and calculate the mean for each label
grouped_data = data.groupby('label')[mfcc_columns].mean()

# Plot bar plots for each label
labels = grouped_data.index
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(25, 15))
axes = axes.flatten()

for i, label in enumerate(labels):
    axes[i].bar(range(1, 21), grouped_data.loc[label], color='teal')
    axes[i].set_title(f'MFCC Means: {label}', fontsize=20)
    axes[i].set_xlabel('MFCC Coefficients', fontsize=20)
    axes[i].set_ylabel('Mean Value', fontsize=20)

print('Genres with a similar tonal structure may have a similar shape across their MFCC coefficients.')
print('As shown for Classical Music: Likely to have more stable and balanced MFCC values, representing smooth transitions and harmonious tones.')
print('Metal/Rock: Shows spikes in certain MFCCs, representing the sharpness and complexity of electric instruments.')

plt.tight_layout()
plt.show()
```

```python
audio_data_metal, sampling_rate_metal = librosa.load('/Users/rajatthakur/MyDocumets/SuperDatascience DataSet/Music ClassifierGTZAN Dataset/genres_original/metal/metal.00000.wav', sr = None )
```

```python
import matplotlib.pyplot as plt
import librosa.display

# Create subplots for side-by-side comparison
plt.figure(figsize=(15, 5))

print("Waveform Comparison")
print("-" * 20)  # Print a line of dashes for visual separation
print("Classical Music:")
print("  - The waveform shows smoother transitions with frequent periods of silence.")
print("  - This reflects the softer and more dynamic nature of classical music.")
print("")
print("Metal Music:")
print("  - The waveform contains loud and frequent peaks with fewer silences.")
print("  - This indicates the intense and energetic style of metal music.")

plt.suptitle('Comparison of Waveforms: Classical vs. Metal', fontsize=16, y=1.02)

# Waveform for classical
plt.subplot(1, 2, 1)
librosa.display.waveshow(audio_data, sr=sampling_rate)
plt.title('Waveform - Classical')
plt.xlabel('Time')
plt.ylabel('Amplitude')
# annotation for classical
plt.text(5, 0.25, 'Smoother transitions\nFrequent silences', fontsize=10,
         bbox=dict(facecolor='white', alpha=0.7))


# Waveform for metal
plt.subplot(1, 2, 2)
librosa.display.waveshow(audio_data_metal, sr=sampling_rate_metal)
plt.title('Waveform - Metal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.text(5, 0.75, 'Louder peaks\nLess silence', fontsize=10,
         bbox=dict(facecolor='white', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
import numpy as np

print('Frequency distribution over time (e.g., classical have smoother transitions, while metal  show higher energy in mid/high frequencies).')
print("### Spectrogram Comparison")
print("-" * 20)  # Print a line of dashes for visual separation
print("- **Classical Music:**")
print("  - Smoother transitions with energy concentrated in lower frequencies.")
print("  - Reflects the softer and more melodic nature of classical music.")
print("")  # Print an empty line for better readability
print("- **Metal Music:**")
print("  - High energy with significant activity in mid and high frequencies.")
print("  - Reflects the aggressive and intense nature of metal music.")

# Compute spectrogram for classical
S_classical = librosa.stft(audio_data) # Short-time Fourier transform
S_db_classical = librosa.amplitude_to_db(np.abs(S_classical), ref=np.max)

# Compute spectrogram for metal
S_metal = librosa.stft(audio_data_metal)
S_db_metal = librosa.amplitude_to_db(np.abs(S_metal), ref=np.max)

# Plot spectrograms side by side
plt.figure(figsize=(15, 5))
plt.suptitle('Spectrogram Comparison: Classical vs. Metal', fontsize=16, y=0.98)


plt.subplot(1, 2, 1)
librosa.display.specshow(S_db_classical, sr=sampling_rate, x_axis='time', y_axis='hz', cmap='coolwarm')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram - Classical')
plt.text(1, 4000, 'Smoother transitions\nDominant in low frequencies', fontsize=10,
         bbox=dict(facecolor='white', alpha=0.8))

plt.subplot(1, 2, 2)
librosa.display.specshow(S_db_metal, sr=sampling_rate_metal, x_axis='time', y_axis='hz', cmap='coolwarm')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram - Metal')
plt.text(1, 8000, 'High energy\nDominant in mid/high frequencies', fontsize=10,
         bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import librosa
import librosa.display

# Create subplots for side-by-side comparison
plt.figure(figsize=(15, 5))

print("### Mel-Spectrogram Insights")
print("-" * 20)  # Print a line of dashes for visual separation
print("- **Classical Music:**")
print("  - Dominates in lower mel-frequencies, representing softer tones.")
print("  - Smoother energy transitions over time.")
print("")  # Print an empty line for better readability
print("- **Metal Music:**")
print("  - Dominates in higher mel-frequencies, representing sharp and aggressive tones.")
print("  - Exhibits sharp bursts of energy, indicative of high intensity.")

# Mel-spectrogram for classical
plt.subplot(1, 2, 1)
plt.suptitle('Mel-Spectrogram Comparison: Classical vs. Metal', fontsize=16, y=0.98)

mel_spectrogram = librosa.feature.melspectrogram(y=audio_data, sr=sampling_rate, n_mels=128)
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
librosa.display.specshow(mel_spectrogram_db, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-Spectrogram - Classical')
plt.text(1, 100, 'Dominates in lower frequencies\nSmooth energy transitions', fontsize=10,
         bbox=dict(facecolor='white', alpha=0.8))

# Mel-spectrogram for metal
plt.subplot(1, 2, 2)
mel_spectrogram_metal = librosa.feature.melspectrogram(y=audio_data_metal, sr=sampling_rate_metal, n_mels=128)
mel_spectrogram_db_metal = librosa.power_to_db(mel_spectrogram_metal, ref=np.max)
librosa.display.specshow(mel_spectrogram_db_metal, sr=sampling_rate_metal, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-Spectrogram - Metal')
plt.text(1, 300, 'High energy in higher frequencies\nSharp bursts of intensity', fontsize=10,
         bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```
